```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# 1. Cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/Wolverinemtb/bank/refs/heads/main/bank.csv")  # o usa la ruta que tengas

# 2. Codificar variables categóricas
df_encoded = df.copy()
le = LabelEncoder()
for col in df_encoded.select_dtypes(include='object'):
    df_encoded[col] = le.fit_transform(df_encoded[col])

# 3. Separar variables independientes (X) y objetivo (y)
X = df_encoded.drop('deposit', axis=1)
y = df_encoded['deposit']

# 4. Escalar variables numéricas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Dividir en train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# 6. Crear y entrenar el modelo
model = LogisticRegression(max_iter=2000)
model.fit(X_train, y_train)

# 7. Predicción y evaluación
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd

# Cargar y procesar datos
df = pd.read_csv("https://raw.githubusercontent.com/Wolverinemtb/bank/refs/heads/main/bank.csv", sep=';')
df = df.iloc[:, 0].str.split(',', expand=True)
df.columns = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome', 'deposit']
df[['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']] = df[['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']].apply(pd.to_numeric)
df['deposit'] = df['deposit'].map({'yes': 1, 'no': 0})
df_encoded = pd.get_dummies(df.drop(columns='deposit'), drop_first=True)
X = df_encoded
y = df['deposit']
```

```python
# Preparar datos
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Modelo, predicción y visualización de matriz de confusión
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Entrenar el modelo
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train_scaled, y_train)

# Predicciones
y_pred = clf.predict(X_test_scaled)
y_pred_proba = clf.predict_proba(X_test_scaled)[:, 1]

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
labels = ['Refused T. Deposits', 'Accepted T. Deposits']

# Visualización
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='rocket_r', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()

# Reporte de clasificación
print("Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=labels))
```

Marketing Insight:
Falsos positivos (FP): El modelo predijo que un cliente sí se suscribiría, pero no lo hizo.
- Esto implica malgastar recursos (llamadas, correos, agentes).

Falsos negativos (FN): El modelo predijo que un cliente no se suscribiría, pero sí lo hizo.
- Aquí estás perdiendo oportunidades de venta.

Objetivo: Minimizar FP para ahorrar presupuesto y FN para no perder ventas.

Precision (Precisión): De los clientes que el modelo dijo que sí, ¿cuántos realmente lo hicieron?

Recall (Sensibilidad): De todos los que realmente sí se suscribieron, ¿cuántos el modelo detectó?

Marketing Insight:
Si se prioriza recall, se llegaría a más posibles suscriptores, pero podrías desperdiciar esfuerzo en muchos que no se suscriben.

Si se prioriza precisión, se contactaría a menos personas, pero con mayor tasa de conversión.

Campañas con recursos limitados → Prioriza precisión.
Campañas agresivas de expansión → Prioriza recall.

```python
# Curvas y gráficas
from sklearn.metrics import roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt
import seaborn as sns

precisions, recalls, _ = precision_recall_curve(y_test, y_pred_proba)
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

importances = clf.feature_importances_
feature_names = X.columns
feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False).head(10)

plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
plt.plot(recalls, precisions, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Tradeoff')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
sns.barplot(data=feature_importances, y='Feature', x='Importance')
plt.title('Top 10 Most Influential Features (Decision Tree)')
plt.tight_layout()
plt.show()
```
